# Peek at the EC silver table

Quick read of the first 9 rows of `nmdc_results.annotation_enzyme_commission`
from the BERDL Delta lakehouse. Run after `ingest_ko_ec_annotations.ipynb`
finishes the EC table (you'll see `_delta_log/00000000000000000000.json`
appear in MinIO when the commit lands).


### Spark Connect

```python
spark = get_spark_session(app_name="peek_ec_silver", tenant_name="nmdc")
print(f"Spark version: {spark.version}")
```

### Preflight: confirm the table is registered

```python
NAMESPACE = "nmdc_results"
TABLE     = "annotation_enzyme_commission"

if spark.sql(f"SHOW TABLES IN {NAMESPACE} LIKE '{TABLE}'").count() == 0:
    raise RuntimeError(
        f"{NAMESPACE}.{TABLE} is not registered in Hive yet. "
        "Wait for ingest_ko_ec_annotations.ipynb (Phase 3) to finish."
    )

print(f"OK: {NAMESPACE}.{TABLE} is registered")
```

### First 9 rows

```python
# Last expression in a cell auto-renders as an HTML table in Jupyter.
spark.sql(f"SELECT * FROM {NAMESPACE}.{TABLE} LIMIT 9").toPandas()
```

### Schema + row count (for context)

```python
n = spark.sql(f"SELECT COUNT(*) AS n FROM {NAMESPACE}.{TABLE}").collect()[0]["n"]
print(f"total rows: {n:,}")

spark.sql(f"DESCRIBE {NAMESPACE}.{TABLE}").toPandas()
```
